```python
import pandas as pd
```

```python
df = pd.read_csv( "ChiTiet_BDS_LocTruong.csv",sep = ";",decimal = ",")

```

```python
df.head(10)
```

```python
df.info()
```

```python
numeric_cols = ["Số phòng ngủ", 'Số phòng tắm, vệ sinh' ]
```

```python
df[numeric_cols] = df[numeric_cols].apply(
    lambda x: x.str.replace("phòng", "", regex=False).astype(float)
)
```

```python
object_cols = ['Nội thất','Pháp lý']
df[object_cols] = df[object_cols].apply(lambda x: x.str.replace(".", "", regex=False))
```

```python
df = df[~df["URL Chi tiết"].str.contains("lahome", na=False)]
```

```python
df = df[~df["Khoảng giá"].str.contains("Thỏa thuận", na=False)]

mask_m2 = df["Khoảng giá"].str.contains("triệu/m²", na=False)
mask_trieu = df["Khoảng giá"].str.contains("triệu", na=False) & ~mask_m2
mask_ty = ~mask_m2 & ~mask_trieu

df["Diện tích"] = (
    df["Diện tích"]
    .str.replace(r"\.(?=\d{3}\b)", "", regex=True)
    .str.replace(",", ".", regex=False)
    .str.extract(r'(\d+\.?\d*)')[0]
    .astype(float)
)

df["Khoảng giá"] = (
    df["Khoảng giá"]
    .str.replace(".", "", regex=False)
    .str.replace(",", ".", regex=False)
)

df.loc[mask_m2, "Khoảng giá"] = (
    df.loc[mask_m2, "Khoảng giá"]
    .str.extract(r'(\d+\.?\d*)')[0]
    .astype(float)
    * df.loc[mask_m2, "Diện tích"]
    / 1000
)

df.loc[mask_trieu, "Khoảng giá"] = (
    df.loc[mask_trieu, "Khoảng giá"]
    .str.extract(r'(\d+\.?\d*)')[0]
    .astype(float)
    / 1000
)

df.loc[mask_ty, "Khoảng giá"] = (
    df.loc[mask_ty, "Khoảng giá"]
    .str.extract(r'(\d+\.?\d*)')[0]
    .astype(float)
)

df["Khoảng giá"] = df["Khoảng giá"].astype(float)
```

```python
df.to_csv(
    "ChiTiet_BDS_LocTruong.csv",
    index=False,
    encoding="utf-8-sig",
    sep=";",
    decimal=",",
    float_format="%.6g"
)
```

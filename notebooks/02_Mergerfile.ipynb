```python
import pandas as pd
import glob
```

```python
input_files = glob.glob('ChiTiet_BDS_LocTruong*.csv')


all_df = []

for file_path in input_files:

    df_temp = pd.read_csv(file_path)


    if 'ID' in df_temp.columns:
        df_temp = df_temp.drop('ID', axis=1)

    all_df.append(df_temp)
    print(f"Đã đọc và xử lý: {file_path}")


if all_df:
    df_final = pd.concat(all_df, ignore_index=True)


    df_final.to_csv('ChiTiet_BDS_LocTruong.csv', index=False, encoding="utf-8-sig")
    print(f"\nThành công! Đã nối {len(all_df)} file vào 'ChiTiet_BDS_LocTruong.csv'")
else:
    print("Không tìm thấy file nào để nối.")
```

```python


file_danh_sach = "ChiTiet_BDS_LocTruong.csv"

try:
    df = pd.read_csv(file_danh_sach)

    if "ID" not in df.columns:
        df.insert(0, "ID", range(1, len(df) + 1))
        df.to_csv(file_danh_sach, index=False, encoding="utf-8-sig")
        print(f"Đã đánh ID thành công cho {len(df)} bản ghi!")

    else:
        print("Cột ID đã tồn tại trong file.")

    display(df.head())
except Exception as e:
    print(f"Lỗi khi đọc file: {e}")
```

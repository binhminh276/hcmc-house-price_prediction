```python
# xử lý PHÁP LÝ
def clean_phaply(x):
    if pd.isna(x):
        return "Không rõ"
    x = str(x).lower()
    if "vi bằng" in x or "vi bang" in x:
        return "Vi bằng / uỷ quyền"
    if "hđmb" in x or "hdmb" in x or "hợp đồng mua bán" in x:
        return "Hợp đồng mua bán"
    if "chờ sổ" in x or "đợi sổ" in x or "sắp nhận sổ" in x:
        return "Đang chờ sổ"
    if "chung" in x or "đồng sở hữu" in x:
        return "Sổ chung"
    if "sổ" in x or "shr" in x:
        return "Sổ riêng"
    return "Không rõ"


df["Pháp lý"] = df["Pháp lý"].apply(clean_phaply)

# xử lý NỘI THẤT
def clean_noithat(x):
    if pd.isna(x):
        return "Không rõ"
    x = str(x).lower()
    if "không nội thất" in x or "nhà trống" in x or "thô" in x:
        return "Không nội thất"
    if "cao cấp" in x or "luxury" in x or "hạng sang" in x or "nhập khẩu" in x:
        return "Nội thất cao cấp"
    if "cơ bản" in x or "ntcb" in x:
        return "Nội thất cơ bản"
    if "full" in x or "đầy đủ" in x:
        return "Full nội thất"
    return "Không rõ"

df["Nội thất"] = df["Nội thất"].apply(clean_noithat)

print(df["Pháp lý"].value_counts())
print(df["Nội thất"].value_counts())

```

```python
import pandas as pd
import numpy as np
import re

# danh sách quận huyện TP.HCM
district_keywords = [
    "Quận 1","Quận 2","Quận 3","Quận 4","Quận 5","Quận 6","Quận 7","Quận 8","Quận 9",
    "Quận 10","Quận 11","Quận 12",
    "Tân Bình","Tân Phú","Bình Thạnh","Gò Vấp","Phú Nhuận","Bình Tân",
    "Thủ Đức",
    "Bình Chánh","Nhà Bè","Củ Chi","Hóc Môn","Cần Giờ"
]

def split_address(addr):

    if pd.isna(addr):
        return pd.Series([None,None,None])

    addr = addr.replace("Hồ Chí Minh","").strip()

    duong = None
    phuong = None
    quan = None

    # tìm phường / xã / thị trấn
    ward_match = re.search(
        r'(Phường\s[^,]+|Xã\s[^,]+|Thị trấn\s[^,]+)',
        addr,
        re.IGNORECASE
    )

    if ward_match:
        phuong = ward_match.group(0).strip()


    # tìm quận huyện theo keyword
    for d in district_keywords:
        if d.lower() in addr.lower():
            quan = d
            break


    # phần còn lại là đường
    duong = addr

    if phuong:
        duong = duong.replace(phuong,"")

    if quan:
        duong = duong.replace(quan,"")

    duong = duong.replace(",", " ").strip()

    if duong == "":
        duong = np.nan


    return pd.Series([duong,phuong,quan])


# áp dụng
df[["Đường","Phường","Quận"]] = df["Địa chỉ chi tiết"].apply(split_address)


# làm sạch cột đường
df["Đường"] = df["Đường"].str.replace(",", "")
df["Đường"] = df["Đường"].str.strip()
df["Đường"] = df["Đường"].replace("", np.nan)
```

```python
import re
import numpy as np

def extract_bedroom(text):
    if pd.isna(text):
        return np.nan
    patterns = [
        r'(\d+)\s*phòng ngủ',
        r'(\d+)\s*pn',
        r'(\d+)\s*phòng'
    ]
    text = text.lower()
    for p in patterns:
        m = re.search(p, text)
        if m:
            return int(m.group(1))
    return np.nan

df["bedroom_desc"] = df["Mô tả"].apply(extract_bedroom)
df["Số phòng ngủ"] = df["Số phòng ngủ"].fillna(df["bedroom_desc"])
df = df.drop(columns=["bedroom_desc"])

def extract_bathroom(text):
    if pd.isna(text):
        return np.nan
    patterns = [
        r'(\d+)\s*toilet',
        r'(\d+)\s*wc',
        r'(\d+)\s*phòng tắm',
        r'(\d+)\s*nhà vệ sinh'
    ]
    text = text.lower()
    for p in patterns:
        m = re.search(p, text)
        if m:
            return int(m.group(1))
    return np.nan

df["bathroom_desc"] = df["Mô tả"].apply(extract_bathroom)
df["Số phòng tắm, vệ sinh"] = df["Số phòng tắm, vệ sinh"].fillna(df["bathroom_desc"])
df = df.drop(columns=["bathroom_desc"])

print(df["Số phòng ngủ"].isna().sum())
print(df["Số phòng tắm, vệ sinh"].isna().sum())
```

```python
def extract_legal(text):
    text = text.lower()
    if "sổ hồng" in text or "shr" in text:
        return "Sổ riêng"
    if "sổ đỏ" in text:
        return "Sổ riêng"
    if "hđmb" in text or "hợp đồng mua bán" in text:
        return "Hợp đồng mua bán"
    if "vi bằng" in text:
        return "Vi bằng / uỷ quyền"
    if "chờ sổ" in text or "đợi sổ" in text or "sắp nhận sổ" in text:
        return "Đang chờ sổ"
    if "chung" in text or "đồng sở hữu" in text:
        return "Sổ chung"

df["phaply_desc"] = df["Mô tả"].apply(extract_legal)
mask = (df["Pháp lý"] == "Không rõ") & (df["phaply_desc"].notna())
df.loc[mask, "Pháp lý"] = df.loc[mask, "phaply_desc"]
df = df.drop(columns=["phaply_desc"])

def extract_furniture(text):
    text = text.lower()
    if "full nội thất" in text or "đầy đủ nội thất" in text:
        return "Full nội thất"
    if "nội thất cao cấp" in text or "luxury" in text or "hạng sang" in text or "nhập khẩu" in text:
        return "Nội thất cao cấp"
    if "nội thất cơ bản" in text:
        return "Nội thất cơ bản"
    if "nhà trống" in text:
        return "Không nội thất"

df["noithat_desc"] = df["Mô tả"].apply(extract_furniture)
mask = (df["Nội thất"] == "Không rõ") & (df["noithat_desc"].notna())
df.loc[mask, "Nội thất"] = df.loc[mask, "noithat_desc"]
df = df.drop(columns=["noithat_desc"])

print((df["Pháp lý"] == "Không rõ").sum())
print((df["Nội thất"] == "Không rõ").sum())
```

```python
df["area_group"] = pd.cut(
    df["Diện tích"],
    bins=[0,40,60,80,120,200,500],
    labels=["<40","40-60","60-80","80-120","120-200",">200"]
)

bedroom_median = df.groupby(
    ["area_group", "Cao tầng"],
    observed=False
)["Số phòng ngủ"].median()

bathroom_median = df.groupby(
    ["area_group", "Cao tầng"],
    observed=False
)["Số phòng tắm, vệ sinh"].median()

def fill_bedroom(row):
    if pd.isna(row["Số phòng ngủ"]):
        return bedroom_median.get(
            (row["area_group"],row["Cao tầng"]),
            np.nan
        )
    return row["Số phòng ngủ"]

df["Số phòng ngủ"] = df.apply(fill_bedroom, axis=1)

def fill_bathroom(row):
    if pd.isna(row["Số phòng tắm, vệ sinh"]):
        return bathroom_median.get(
            (row["area_group"],row["Cao tầng"]),
            np.nan
        )
    return row["Số phòng tắm, vệ sinh"]

df["Số phòng tắm, vệ sinh"] = df.apply(fill_bathroom, axis=1)

print(bedroom_median)
print(bathroom_median)

print(df["Số phòng ngủ"].isna().sum())
print(df["Số phòng tắm, vệ sinh"].isna().sum())
```
